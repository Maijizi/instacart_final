```python
num_bins = 50

fig, ax = plt.subplots()

x=df_orders.groupby(['user_id'])['order_number'].count()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins)

ax.set_xlabel('Number of orders per user')
ax.set_ylabel('Frequency')
ax.set_title(r'Histogram of Distribution of Orders per User')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```

```python
N_points = 100000
n_bins = 20

# Generate a normal distribution, center at x=0 and y=5
x = df_orders.groupby(['user_id'])['user_id']
y = df_orders.groupby(['user_id'])['order_number'].count()

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the `bins` kwarg
axs[0].hist(x, bins=n_bins)
axs[1].hist(y, bins=n_bins)
```
