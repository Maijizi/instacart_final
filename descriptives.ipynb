# Data Wrangling for PCA

```python
aisle_df = df_orders.merge(df_order_products_train, on='order_id')
aisle_df = aisle_df.merge(df_products, on='product_id')
aisle_df = aisle_df.merge(df_aisle, on='aisle_id')
```

```python
aisle_df.tail(50)
```

```python
aisle_df = aisle_df[['user_id', 'aisle']]
```

```python
aisle_df.tail(10)
```

```python
aisle_dummy_df = pd.get_dummies(aisle_df.aisle)
aisle_dummy_df = pd.concat([aisle_df, aisle_dummy_df], axis=1)
aisle_dummy_df = aisle_dummy_df.drop(['aisle'], axis=1)
aisle_dummy_df
```

```python
aisle_group = aisle_dummy_df.groupby('user_id')
aisle_group.head(100)
```

```python
user_aisle_df = aisle_group.sum()
user_aisle_df
```

# PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
```

```python
#Fitting the PCA algorithm with our Data
pca = PCA().fit(user_aisle_df)

#Plotting the Cumulative Summation of the Explained Variance
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Pulsar Dataset Explained Variance')
plt.show()
```

```python
pca = PCA(n_components=60)
pca.fit(user_aisle_df)
pca_samples = pca.transform(user_aisle_df)

ps = pd.DataFrame(pca_samples)
ps.head()
```

```python
tocluster = pd.DataFrame(ps[[5,37]])
print (tocluster.shape)
print (tocluster.head())

fig = plt.figure(figsize=(8,8))
plt.plot(tocluster[5], tocluster[37], 'o', markersize=2, color='blue', alpha=0.5, label='class1')

plt.xlabel('x_values')
plt.ylabel('y_values')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

clusterer = KMeans(n_clusters=5,random_state=42).fit(tocluster)
centers = clusterer.cluster_centers_
c_preds = clusterer.predict(tocluster)
print(centers)
```

```python
import matplotlib
fig = plt.figure(figsize=(8,8))
colors = ['orange','blue','purple','green','pink']
colored = [colors[k] for k in c_preds]
print (colored[0:10])
plt.scatter(tocluster[5],tocluster[37],  color = colored)
for ci,c in enumerate(centers):
    plt.plot(c[0], c[1], 'o', markersize=10, color='red', alpha=0.9, label=''+str(ci))

plt.xlabel('x_values')
plt.ylabel('y_values')
plt.legend()
plt.show()
```
